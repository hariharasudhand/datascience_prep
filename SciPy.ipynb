## Mathematical Constants


```python
from scipy.constants import pi,golden_ratio

print(f" Pi = {pi}")
print(f"Golden Constants = {golden_ratio}")

#Find area of a Circle
radius = 5

print(f" Area of the circle is = {pi*radius*radius}")
```

# List all Scipy Constants

```python
from scipy import constants

constants_list = dir(constants)

print(f" List of Constants \n ")
#print(constants_list)

for constant, description in constants.physical_constants.items():
    print(f"{constant}: {description[0]} (Value: {description[1]}, Units: {description[2]})")
```
